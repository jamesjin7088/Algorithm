## 문제 1: 다양한 괄호 검사기 (스택 응용)

### 배경
프로그래밍 언어에서는 괄호의 짝이 올바르게 맞는지 확인하는 것이 중요합니다. 이 문제에서는 여러 종류의 괄호 '()', '[]', '{}'가 올바르게 짝지어졌는지 확인하는 프로그램을 작성해봅시다.

### 문제
입력으로 주어진 문자열에 포함된 괄호들이 올바르게 짝지어졌는지 확인하는 프로그램을 작성하세요. 올바른 괄호 표현식의 조건은 다음과 같습니다:
- 같은 종류의 괄호가 짝을 이루어야 합니다 (예: '(' 와 ')', '[' 와 ']', '{' 와 '}').
- 서로 다른 종류의 괄호 쌍은 서로를 교차하지 않아야 합니다 (예: '[{]}' 는 올바르지 않음).
- 모든 열린 괄호는 반드시 닫혀야 합니다.

### 입력
- 괄호들로 이루어진 여러 문자열

### 출력
- 각 문자열이 올바른 괄호 표현식이면 "올바른", 아니면 "올바르지 않은"을 출력합니다.

```python

## 문제 1: 다양한 괄호 검사기 (스택 응용)

# 배경
# 프로그래밍 언어에서는 괄호의 짝이 올바르게 맞는지 확인하는 것이 중요합니다.
# 이 문제에서는 여러 종류의 괄호 '()', '[]', '{}'가 올바르게 짝지어졌는지 확인하는 프로그램을 작성해봅시다.

# 문제
# 입력으로 주어진 문자열에 포함된 괄호들이 올바르게 짝지어졌는지 확인하는 프로그램을 작성하세요.
# 올바른 괄호 표현식의 조건은 다음과 같습니다:
# - 같은 종류의 괄호가 짝을 이루어야 합니다 (예: '(' 와 ')', '[' 와 ']', '{' 와 '}').
# - 서로 다른 종류의 괄호 쌍은 서로를 교차하지 않아야 합니다 (예: '[{]}' 는 올바르지 않음).
# - 모든 열린 괄호는 반드시 닫혀야 합니다.

class BracketStack:
    def __init__(self):
        # 스택 초기화
        self.stack = []

    def push(self, item):
        # 스택에 아이템 추가
        self.stack.append(item)

    def pop(self):
        # 스택이 비어있지 않으면 맨 위 아이템 제거 및 반환
        if not self.is_empty():
            return self.stack.pop()
        return None

    def is_empty(self):
        # 스택이 비어있는지 확인
        return len(self.stack) == 0

    def peek(self):
        # 스택이 비어있지 않으면 맨 위 아이템 반환 (제거하지 않음)
        if not self.is_empty():
            return self.stack[-1]
        return None

def check_brackets(expression):
    # 여기에 괄호 검사 코드를 작성하세요
    # 1. 여는 괄호가 나오면 스택에 추가
    # 2. 닫는 괄호가 나오면 스택에서 제거하고 짝이 맞는지 확인
    # 3. 모든 검사 후 스택이 비어있으면 올바른 괄호 표현식
    brackets= {')':'(', ']' : '[', '}' : '{'}
    stack = BracketStack()
    for i in expression:
        if i in brackets.values():
            stack.push(i)
        elif i in brackets:
            a = stack.pop()
            if not a or a != brackets[i]: return False
    return True if stack.is_empty() else False


# 테스트 케이스
test_expressions = [
    "()",
    "()[]{}",
    "([{}])",
    "([)]",
    "({[]})",
    ")("
]

for expression in test_expressions:
    result = check_brackets(expression)
    print(f"'{expression}' 는 {'올바른' if result else '올바르지 않은'} 괄호 표현식입니다.")



```

## 문제 2: 원형 카드 게임 (조세퍼스 문제 - 큐 응용)

### 배경
N명의 사람이 원형으로 앉아 있고, 각자 번호가 1부터 N까지 부여되어 있습니다. 이제 K번째 사람을 계속해서 제거하는 게임을 진행합니다. 한 사람이 제거되면, 다음 사람부터 다시 K번째 사람을 세어 제거합니다. 이 과정은 마지막 한 사람이 남을 때까지 계속됩니다.

### 문제
N과 K가 주어졌을 때, 제거되는 사람들의 순서와 마지막으로 남는 사람을 구하는 프로그램을 작성하세요.

### 입력
- 첫 줄에 사람의 수 N(1 ≤ N ≤ 100)과 제거할 번호 K(1 ≤ K ≤ 10)가 주어집니다.

### 출력
- 첫 줄에는 제거되는 사람들의 순서를 공백으로 구분하여 출력합니다.
- 두 번째 줄에는 마지막으로 남는 사람의 번호를 출력합니다.

```python
## 문제 2: 원형 카드 게임 (조세퍼스 문제 - 큐 응용)

# 배경
# N명의 사람이 원형으로 앉아 있고, 각자 번호가 1부터 N까지 부여되어 있습니다.
# 이제 K번째 사람을 계속해서 제거하는 게임을 진행합니다.
# 한 사람이 제거되면, 다음 사람부터 다시 K번째 사람을 세어 제거합니다.
# 이 과정은 마지막 한 사람이 남을 때까지 계속됩니다.

# 문제
# N과 K가 주어졌을 때, 제거되는 사람들의 순서와 마지막으로 남는 사람을 구하는 프로그램을 작성하세요.

class CircularQueue:
    def __init__(self, size):
        self.size = size
        self.queue = [None] * size
        self.front = self.rear = 0

    def enqueue(self, item):
        if self.is_full():
            print("Queue is full")
            return False

        self.rear = (self.rear + 1) % self.size
        self.queue[self.rear] = item
        return True

    def dequeue(self):
        if self.is_empty():
            print("Queue is empty")
            return None

        self.front = (self.front + 1) % self.size
        item = self.queue[self.front]
        self.queue[self.front] = None
        return item

    def is_empty(self):
        return self.front == self.rear

    def is_full(self):
        return (self.rear + 1) % self.size == self.front

    def peek(self):
        if self.is_empty():
            return None
        return self.queue[(self.front + 1) % self.size]

    def queue_len(self):
        if self.rear >= self.front:
            return self.rear - self.front
        else:
            return self.size - (self.front - self.rear)

def josephus_problem(n, k):
    # 여기에 원형 카드 게임(조세퍼스 문제) 시뮬레이션 코드를 작성하세요
    # n: 사람의 수
    # k: 제거할 번호

    # 구현해야 할 내용:
    # 1. 1부터 n까지의 사람들을 큐에 추가합니다.
    # 2. k-1명은 건너뛰고 k번째 사람을 제거합니다.
    # 3. 제거된 사람의 번호를 기록합니다.
    # 4. 한 사람이 남을 때까지 2-3을 반복합니다.
    # 5. 제거된 순서와 마지막 남은 사람을 반환합니다.
    q = CircularQueue(n+1)
    for i in range(1, n + 1):
        q.enqueue(i)
    removed_order = []
    while q.queue_len() > 1:
        for j in range(k-1):
            q.enqueue(q.dequeue())
        removed_person = q.dequeue()
        removed_order.append(removed_person)
    last_person = q.dequeue()
    return removed_order, last_person  




# 테스트 코드
N, K = map(int, input("사람의 수와 제거할 번호를 입력하세요 (예: 7 3): ").split())
removed_order, last_person = josephus_problem(N, K)
print("제거되는 순서:", " ".join(map(str, removed_order)))
print("마지막으로 남는 사람:", last_person)

# 테스트 케이스 예시
# 입력: 7 3
#
# 출력:
# 제거되는 순서: 3 6 2 7 5 1
# 마지막으로 남는 사람: 4
#
# 설명:
# 1. 처음 상태: [1, 2, 3, 4, 5, 6, 7]
# 2. 3번째 사람인 3을 제거: [1, 2, 4, 5, 6, 7]
# 3. 다음으로 3번째 사람인 6을 제거: [1, 2, 4, 5, 7]
# 4. 다음으로 3번째 사람인 2를 제거: [1, 4, 5, 7]
# 5. 다음으로 3번째 사람인 7을 제거: [1, 4, 5]
# 6. 다음으로 3번째 사람인 5를 제거: [1, 4]
# 7. 다음으로 3번째 사람인 1을 제거: [4]
# 8. 마지막으로 4가 남게 됩니다.
```
